```python
%matplotlib inline

# Import dependencies.
import matplotlib.pyplot as plt

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.plot(x_axis, y_axis)

# or
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()


# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)

# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
# (Only use plt.show once per session, may have backlash and might need to 'Rrestart and CLear Output' of kernel / don't mix with other graphs)

```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()

plt.figure(figsize=(20, 10))
```

```python


import plotly.graph_objects as go

# Generate example data
import numpy as np

x = x_axis
y = y_axis

# Build figure
fig = go.Figure()

# Add scatter trace with medium sized markers
fig.add_trace(
    go.Scatter(
        mode='markers',
        x=y,
        y=x,
        marker=dict(
            color='Blue',
            size=20,
            opacity=0.2,
            line=dict(
                color='black',
                width=2
            )
        ),
        showlegend=True
    )
)




fig.show()




```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)


fig, ax = plt.subplots()
for color in ['Boston']:  
    x, y = (y_axis, x_axis)
    color = 'blue'
    scale = [i*5 for i in y_axis]
    ax.scatter(x, y, c=color, s=scale, label='Boston',
               alpha=0.20, edgecolors='black')


ax.grid(False)
plt.xlabel("Fare($)")
plt.ylabel("Date")
ax.legend()
plt.xlim(0, 50)
plt.gca().invert_yaxis()


plt.show
```
